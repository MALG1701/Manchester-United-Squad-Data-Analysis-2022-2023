```python
import matplotlib.pyplot as plt

a=plantilla_mufc['xG 90']
b=plantilla_mufc['xAG 90']
#data_squad['Avg xG+xAG']=data_squad['xG+xAG'].mean()
c=plantilla_mufc['Apellido']
c_t=c.transpose()


plt.figure(figsize=(15,5))
plt.xticks(rotation=0)
plt.title('xG-xAG Manchester United temporada 2022-2023')
plt.xlabel('Goles Esperados (xG)')
plt.ylabel('Asistencias Esperadas (xAG)')
plt.scatter(a,b,label='test')
for i, txt in enumerate(c):
    plt.annotate(txt, (a[i], b[i]))
```
